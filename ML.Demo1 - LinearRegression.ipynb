- Biểu diễn trực quan dữ liệu heights, weights

```python
## Đọc tập tin chiều cao heights.txt vào array 'heights' và chuyển đổi đơn vị (mét)
f    = open(folder + '/Data/heights.txt')
data = f.read()
f.close()

heights = np.array(eval(data)) * 0.0254
print('heights[] (mét):', heights.size, 'mẫu;', heights[0:5])
```

### <i>Phương pháp <b>Gradient Descent</b></i>
---

```python
## Mô hình hóa dữ liệu
##    Lưu ý: Thêm 1 vào cột đầu tiên của X để có cùng số chiều với vector tham số (m, b)
X    = heights
xHat = np.c_[np.ones(heights.size), X]
y    = weights
print(xHat)
```

```python
##------------------------------------------------------------------------------
## Hàm thực hiện Gradient Descent (Tính đạo hàm bằng cách nhân ma trận)
##------------------------------------------------------------------------------
def gradient_descent(rho, xHat, y, numIterations):
    #     rho:  learning rate
    #     xHat: ma trận input  [[1 x0],[1 x1], [1  x2], ...]
    #     y:    vector  output [y0, y1, y2, ...]
    #     numIterations: số vòng lặp

    # Kích thước tập huấn luyện
    N = xHat.shape[0]

    # Khởi tạo (tùy ý) vector tham số cần xác định: theta = (b, m) = [1, 1]
    theta = np.ones(2)

    # Tìm cực tiểu địa phương của loss fuction (di chuyển dần dần ngược dấu đạo hàm)
    for iter in range(0, numIterations):
        yHat = xHat @ theta
        loss = yHat - y

        # Đạo hàm của hàm L (riêng phần theo theta[0] và theta[1])
        gradient = (xHat.T @ loss) / N

        # Di chuyển ngược dấu đạo hàm đến điểm mới
        theta = theta - rho * gradient
    return theta
```

```python
## Áp dụng Gradient Descent để xác định theta = (b, m)
rho   = 0.001
theta = gradient_descent(rho, xHat, y, 10000)
print(theta)
```

```python
## Phương trình đường thẳng (không gian 2D): y = m.x + b
weights_predict = theta[1] * X + theta[0]
print('Phương trình hồi quy tuyến tính:')
print('   weight = (', theta[1], ') * height + (', theta[0], ')')
```

```python
## Biểu diễn trực quan kết quả của Gradient Descent
plt.xlim(1.6, 2.2)
plt.ylim(60, 150)
plt.scatter(xHat[:,1], weights, color = "green")
plt.plot(X, weights_predict)
plt.xlabel("heights")
plt.ylabel("weights")
plt.show()
```

### <i>Sử dụng thư viện <b>sklearn</b></i>
---

```python
## Sử dụng thư viện sklearn với X có kiểu np.array
regr = LinearRegression(fit_intercept = False) # fit_intercept = False ---> Tính toán bias
regr.fit(xHat, y) # huấn luyện mô hình
regr.coef_
```

```python
## Phương trình đường thẳng (không gian 2D): y = m.x + b
slope = regr.coef_.flatten()[1]
bias  = regr.coef_.flatten()[0]

weights_predict_sklearn = slope * X + bias

print('Phương trình hồi quy tuyến tính:')
print('   weight = (', slope, ') * height + (', bias, ')')
```

```python
## Biểu diễn trực quan kết quả của hàm sklearn
plt.xlim(1.6, 2.2)
plt.ylim(60, 150)
plt.scatter(xHat[:,1], weights, color = "blue")
plt.plot(X, weights_predict_sklearn)
plt.xlabel("Heights")
plt.ylabel("Weights")
plt.show()
```

```python
## Sử dụng regr.predict(X).predict() thay vì áp dụng công thức hồi quy
plt.xlim(1.6, 2.2)
plt.ylim(60, 150)
plt.scatter(xHat[:,1], weights)
plt.plot(X, regr.predict(xHat), color = 'red')
plt.xlabel("Heights")
plt.ylabel("Weights")
plt.show()
```

```python
## Sử dụng thư viện sklearn với X có kiểu dataframe
model = LinearRegression()

## Huấn luyện mô hình
model.fit(X = pd.DataFrame(X), y = y)
```

```python
## Phương trình đường thẳng (không gian 2D): y = m.x + b
slope = model.coef_[0]
bias  = model.intercept_

print('Phương trình hồi quy tuyến tính:')
print('   weight = (', slope, ') * height + (', bias, ')')
```

---
### <i>Phương pháp dựa trên <b>ma trận giả nghịch đảo (pseudo-inverse matrix)</b></i>
---

```python
## Phương pháp dựa trên ma trận giả nghịch đảo
A = xHat.T @ xHat
b = xHat.T @ y

## Vector tham số
w = np.linalg.pinv(A) @ b

slope3 = w[1]
bias3  = w[0]

print('Phương trình hồi quy tuyến tính:')
print('   weight = (', slope3, ') * height + (', bias3, ')')
```

## 1b. **Multiple Linear Regression - (mtcars.csv)**
---

```python
## Mở tập tin dữ liệu
data = pd.read_csv(folder + '/Data/mtcars.csv')
print(data.head())
```

```python
## Quan sát dữ liệu
data.hist()
plt.show()
```

```python
## Mô hình hóa đa biến
model = LinearRegression()

## Huấn luyện mô hình
model.fit(X = data.loc[:, ['wt', 'hp','disp']], y = data['mpg'])
```

```python
## Hiển thị kết quả các tham số:
print('Intercept w_0 =', model.intercept_)
print('          w_i =', model.coef_)
```

---
### <i>Đánh giá mô hình</i>
---

```python
## Mô hình hóa dữ liệu
X = data.wt
y = data.mpg

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 1)
```

```python
## Huấn luyện mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X = pd.DataFrame(X_train), y = y_train)
```

```python
## Kiểm thử mô hình
y_pred = model.predict(pd.DataFrame(X_test))
```

```python
## Hàm tính toán các chỉ số MAE, MSE
def regression_eval(y_test, y_pred):
    mae  = mean_absolute_error(y_pred, y_test)
    mse  = mean_squared_error(y_pred, y_test)
    return mae, mse
```

```python
## Đánh giá mô hình hồi quy bằng các chỉ số
errors   = regression_eval(y_test, y_pred)
r2_total = model.score(pd.DataFrame(X), y)
r2_train = model.score(pd.DataFrame(X_train), y_train)
r2_test  = model.score(pd.DataFrame(X_test),  y_test)

## Hiển thị giá trị các chỉ số
print(f'MAE = {errors[0]:.1f}%')
print(f'MSE = {errors[1]:.1f}%')
print(f'R2_Total giải thích được {(r2_total * 100):.1f}% toàn bộ dữ liệu')
print(f'R2_Train giải thích được {(r2_train * 100):.1f}% training set')
print(f'R2_Test  giải thích được {(r2_test  * 100):.1f}% test set')
```

## 2. **Simple Linear Regression**
---

```python
## Công thức hiệp phương sai
mean_x  = np.mean(heights)
delta_x = (heights - mean_x).flatten()

mean_y  = np.mean(weights)
delta_y = (weights - mean_y).flatten()

numerator   = delta_x @ delta_y.T
denominator = delta_x @ delta_x.T

## Vector tham số
slope4  = numerator/denominator
bias4   = mean_y - (slope4 * mean_x)

print('Phương trình hồi quy tuyến tính:')
print('   weight = (', slope4, ') * height + (', bias4, ')\n')
```
